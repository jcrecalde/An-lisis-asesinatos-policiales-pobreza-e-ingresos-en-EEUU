```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

deaths_by_police es el csv transformado de Deaths_by_Police.csv.

```python
deaths_by_police = pd.read_csv("datasets\deaths_by_police(modificado).csv", encoding="ISO-8859-1")  
deaths_by_police.head() 
```

```python
deaths_by_police.describe()
```

Distribución de edades

```python
deaths_per_year = deaths_by_police['year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=deaths_per_year.index, y=deaths_per_year.values, palette="viridis")
plt.title("Número de muertes por año")
plt.xlabel("Año")
plt.ylabel("Número de muertes")
plt.show()
```

Este gráfico muestra el número de muertes por año. En 2015, se registró el mayor número de muertes, seguido de 2016, que también tuvo una cifra alta. En 2017, hubo un descenso significativo, posiblemente debido a las medidas de seguridad implementadas en respuesta a los altos números de años anteriores.

```python
deaths_per_gender = deaths_by_police["gender"].value_counts()  

plt.figure(figsize=(6, 4)) 
sns.barplot(x=deaths_per_gender.index, y=deaths_per_gender.values, palette="viridis")  
plt.title("Numero de muertes por genero")
plt.xlabel("Genero") 
plt.ylabel("Numero de muertes")
plt.show()
```

Como se puede ver en el grafico la mayoria de las muertes son Masculinas

```python
deaths_per_race = deaths_by_police["race"].value_counts() 

plt.figure(figsize=(10, 6)) 
sns.histplot(deaths_by_police['race'], bins=len(deaths_per_race), kde=False, palette='viridis')
plt.title("Muertes por genero") 
plt.xlabel("Raza") 
plt.ylabel("Numero de muertes") 
plt.show()
```

Las letras significan: W(blancos), B(negros), H(hispanicos), A(asiaticos), N(nativos), O(otros)

La verdad es que no era lo que esperaba. En un país donde prevalece la discriminación hacia las personas de color, sorprende la gran diferencia en el número de muertes entre personas blancas y negras.

```python
deaths_by_state = deaths_by_police['state'].value_counts().sort_index()

data_matrix = deaths_by_state.values.reshape(-1, 1)

state_labels = deaths_by_state.index

plt.figure(figsize=(15, 8))
sns.heatmap(data_matrix, annot=True, fmt="d", cmap="viridis", yticklabels=state_labels)
plt.title("Número de muertes por estado")
plt.xlabel("Número de muertes")
plt.ylabel("Estado")
plt.show()
```

El estado con el mayor número de muertes registradas es California (CA). Mi conclusión es que esto se debe a que es el estado más poblado de Estados Unidos. Le siguen Texas y Florida.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='manner_of_death', y='age', data=deaths_by_police, palette='viridis')
plt.title('Distribución de edad por manera de muerte')
plt.xlabel('Manera de muerte')
plt.ylabel('Edad')
plt.show()
```

Como se puede ver en el grafico las muertes solamente por dispario tienen un indice mas alto.

```python
deaths_by_mental_illnes = deaths_by_police["signs_of_mental_illness"].value_counts() 
 
plt.figure(figsize=(6, 6))
deaths_by_mental_illnes.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(deaths_by_mental_illnes)))
plt.title('Número de muertes por señal de enfermedad mental')
plt.ylabel('')  
plt.show()
```

En el gráfico de pastel, se observa que solo el 25% de las muertes son causadas por personas con una enfermedad mental.

```python
deaths_by_body_camera = deaths_by_police["body_camera"].value_counts() 
 
plt.figure(figsize=(6, 4)) 
sns.barplot(x=deaths_by_body_camera.index, y=deaths_by_body_camera.values, palette="viridis") 
plt.title("Numero de muertes por uso de cámaras corporales") 
plt.xlabel("Camara corporal") 
plt.ylabel("Número de muertes") 
plt.xticks([0, 1], ["No", "Si"]) 
plt.show()
```

Este gráfico sugiere que la presencia de una cámara de seguridad podría reducir la probabilidad de un siniestro de asesinato

```python
df_info = pd.read_csv("datasets\inf_eeuu(modificado).csv") 
```

```python
plt.figure(figsize=(25, 12))
sns.scatterplot(x='percent_completed_hs', y='poverty_rate', data=df_info)
plt.title('Tasa de Pobreza vs. Porcentaje de Completado de Secundaria')
plt.xlabel('Porcentaje Completado de Secundaria')
plt.ylabel('Tasa de Pobreza')
plt.show()
```

Como se puede notar, la mayoría de los puntos se concentran en el eje X entre el 60% y el 100%, y en el eje Y entre 0 y 40%. Esto sugiere que aquellos que completan la educación secundaria tienen una menor probabilidad de caer en niveles de pobreza. Sin embargo, existen numerosos casos en los que esto no ocurre debido a diversas circunstancias, como la calidad de la educación recibida, la debilidad del mercado laboral, factores culturales o la persistencia de dificultades a lo largo de varias generaciones en la familia para romper este ciclo

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='geographic_area', y='median_income', data=df_info)
plt.title('Distribución del Ingreso Mediano por Región Geográfica')
plt.xlabel('Región Geográfica')
plt.ylabel('Ingreso Mediano')
plt.xticks(rotation=90)
plt.show()
```

- Las regiones muestran una variabilidad significativa en los ingresos medianos. Algunas regiones tienen cajas más altas, lo que indica una mayor variabilidad en los ingresos.  

- Algunas regiones tienen medianas de ingreso mediano más altas. Por ejemplo, regiones como Connecticut (CT) y Maryland (MD) tienen medianas relativamente altas. 

- Hay bastantes valores atípicos en casi todas las regiones. Esto indica que en cada región hay ciudades o áreas con ingresos significativamente más altos o más bajos que la mayoría.

- Algunas regiones, como Alaska (AK) y Dakota del Norte (ND), tienen rangos intercuartiles más estrechos, lo que sugiere una menor variabilidad en los ingresos medianos dentro de esas regiones.

```python
df_numeric = df_info.select_dtypes(include='number')

# Crear el mapa de calor de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor de Correlación")
plt.show()
```

Lo primero que puedo notar es que la diagonal principal tiene todos los valores en 1 lo que quiere decir que cada variable esta perfectamente correlacionada consigo misma. 

share_white_x y share_black_x: -0.68. ---> Indica que a medida que aumente el porcentaje de la poblacion blanca, dismimuye el porcentaje de poblacion negra.
share_white_x y poverty_rate: -0.35. ---> Esto sugiere que un mayor porcentaje de población blanca está asociado con una menor tasa de pobreza. 
share_black_x y poverty_rate: 0.25. ----> Esto indica que un mayor porcentaje de población negra está asociado con una mayor tasa de pobreza.  
share_hispanic_y y median_income: -0.06. ---->  Esta correlación es bastante débil, lo que sugiere poca relación entre el porcentaje de población hispana y el ingreso medio.  
percent_completed_hs y poverty_rate: -0.59 ---> Una fuerte correlación negativa, lo que indica que a medida que aumenta el porcentaje de personas que completan la secundaria, la tasa de pobreza disminuye.  
 
share_native_american_x: Correlaciones bajas con la mayoría de las variables. Lo que sugiere que no hay una relación fuerte entre el porcentaje de nativos americanos y las otras variables en tu conjunto de datos. 
median_income y percent_completed_hs: 0.48. ----> Indica que un mayor ingreso medio está asociado con un mayor porcentaje de personas que completan la secundaria. 
median_income y poverty_rate: -0.59. ---> A medida que aumenta el ingreso medio, la tasa de pobreza disminuye. 


Conclusiones: La educación(porcentaje de personas que completan la secundaria) tiene una fuerte relación inversa con la pobreza. Indica que los paises deben enfocar politicas enfocadas a la educacion ya que produce una reducción de la pobreza. 
También, el ingreso medio esta inversamente relacionado con la pobreza y positivamente relacionado con la educación, lo que indica la importancia de ingresos medios para mejorar la calidad de vida y a la reducción de pobreza.

```python
unique_regions = df_info['geographic_area'].unique()  

plt.figure(figsize=(15, 10))
for region in unique_regions:
    sns.kdeplot(df_info[df_info['geographic_area'] == region]['median_income'], label=region, fill=True)
plt.title('Distribución de Ingresos Medianos por Región Geográfica')
plt.xlabel('Ingresos Medianos')
plt.ylabel('Densidad')
plt.xlim(0, df_info['median_income'].max())
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

- La mayoría de las regiones tienen ingresos medianos concentrados en niveles más bajos.
- Hay variación en la dispersión de los ingresos medianos entre las regiones.
- Algunas regiones tienen ingresos medianos significativamente más altos, lo cual es visible por las colas de las curvas que se extienden hacia la derecha.

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='median_income', y='poverty_rate', data=df_info)
plt.title('Ingresos Medianos vs. Tasa de Pobreza')
plt.xlabel('Ingresos Medianos')
plt.ylabel('Tasa de Pobreza')
plt.show()
```

Este grafico muestra que cuanto menos ingresos tenes el indice de pobreza aumenta.(Algo logico!)
