Elimino las filas que continen valores NaN en: median_household_income_2015

```python
median_household_income_2015.dropna(inplace=True)
```

Renombro las columnas y elimino los espacios.

```python
median_household_income_2015.rename(columns={"Geographic Area": "geographic_area", "City": "city", "Median Income": "median_income"}, inplace=True) 
pct_over_25_completed_high_school.rename(columns={"Geographic Area": "geographic_area", "City":"city"}, inplace=True) 
pct_people_below_poverty_level.rename(columns={"Geographic Area": "geographic_area", "City": "city"}, inplace=True) 
share_of_race_by_city.rename(columns={"Geographic area": "geographic_area", "City": "city"}, inplace=True)
```

```python
def clean_city_names(df): 
    df["city"] = df["city"].str.strip().str.lower() 
    return df
```

```python
median_household_income_2015 = clean_city_names(median_household_income_2015) 
pct_over_25_completed_high_school = clean_city_names(pct_over_25_completed_high_school) 
pct_people_below_poverty_level = clean_city_names(pct_people_below_poverty_level) 
share_of_race_by_city = clean_city_names(share_of_race_by_city)
```

```python
merged_df = share_of_race_by_city.merge(share_of_race_by_city, on=["geographic_area", "city"], how="left") 
merged_df = merged_df.merge(median_household_income_2015, on=["geographic_area", "city"], how="left") 
merged_df = merged_df.merge(pct_over_25_completed_high_school, on=["geographic_area", "city"], how="left") 
merged_df = merged_df.merge(pct_people_below_poverty_level, on=["geographic_area", "city"], how="left")
```

Cambio el delimitador de los números decimales de punto a coma para luego trabajar en Power BI

```python
dfinfo = pd.read_csv("datasets\inf_eeuu(modificado).csv") 
dfinfo
```

```python
def replace_decimal(x):
    if isinstance(x, (float, int)):
        return str(x).replace('.', ',')
    return x
```

```python
for col in dfinfo.select_dtypes(include=['float64', 'int64']).columns:
    dfinfo[col] = dfinfo[col].apply(replace_decimal)
```

Por eso, voy a enfocar mi análisis en las áreas geográficas con mayor concentración de personas en situación de pobreza. Mi objetivo es comprender las razones detrás de este nivel de pobreza. Para ello, voy a generar un nuevo archivo CSV específicamente para estas áreas.

```python
df_data_top4area = pd.read_csv("datasets\inf_eeuu.csv")
```

```python
df_data_top4area
```

```python
df_data_top4area.dtypes
```

```python
area_selec = ["TX", "CA", "PA", "NO"] 
columns_relevant = ["city", "geographic_area", "share_white_x", "share_black_x", "share_native_american_x", "share_asian_x", "share_hispanic_x", "share_white_y", "share_black_y", 
                    "share_native_american_y", "share_asian_y", "share_hispanic_y", "median_income", "percent_completed_hs", "poverty_rate"] 
```

```python
data_filter = df_data_top4area.loc[df_data_top4area["geographic_area"].isin(area_selec), columns_relevant]
```

```python
data_filter.reset_index(drop=True, inplace=True)
```

```python
columns_convert = ["share_white_x", "share_black_x", "share_native_american_x", "share_asian_x", "share_hispanic_x", "share_white_y", "share_black_y", 
                    "share_native_american_y", "share_asian_y", "share_hispanic_y", "median_income", "percent_completed_hs", "poverty_rate"]
```

```python
for column in columns_convert:  
    data_filter[column] = pd.to_numeric(data_filter[column].str.replace(",", "."), errors="coerce") 
```

```python
for column in columns_convert: 
    data_filter[column] = data_filter[column].apply(lambda x: f"{x:.2f}".replace('.', ',') if pd.notnull(x) else '')
```
